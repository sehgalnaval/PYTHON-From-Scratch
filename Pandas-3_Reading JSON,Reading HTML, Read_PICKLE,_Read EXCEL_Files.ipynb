```python
# Use Selenium:
# Firstly in cmd install selenium and if python is not installed install it too and add it to your environmental variable and then
# after installing python in cmd write: pip install selenium webdriver-manager
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
# This script sets up Selenium to run in headless mode (no visible browser window), retrieves the HTML content, and uses pandas to parse the tables.
```

```python
# Initialize WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

# Open the page
url = 'https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/'
driver.get(url)
```

```python
 # Get page source
html_content = driver.page_source
```

```python
# Close the WebDriver
driver.quit()
```

```python
# I used the StringIO Method because the warning showed:pandas.read_html() will soon require the 
# HTML content to be wrapped in a StringIO object rather than being passed directly as a string.
```

```python
# Wrap the HTML content in StringIO
html_io = StringIO(html_content)
```

```python
# Parse HTML with pandas
dfs = pd.read_html(html_io)
```
