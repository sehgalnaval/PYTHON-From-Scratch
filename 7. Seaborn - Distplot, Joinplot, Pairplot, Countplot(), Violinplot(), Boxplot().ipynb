#### Univariate analysis focuses on understanding individual variables in a dataset, like only 1 feature. - Bivariate analysis examines relationships between two variables, that is relation between two features in a dataset. - Multivariate analysis deals with the interactions and relationships among three or more variables. This type of analysis which involve 3d or more type is made us understood using the Seaborn library

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Above we want to create a model that is able to predict what tip will be given using the other features of the dataset, so the tip is the 
# dependent feature now and all others are Independent features.

# We will use Correlation to find out the inter-relationshiop between the features(!Note the features should be only integer and float,
#as it does not workout for categorical features, as they are object)
```

### Correlation with Heatmap
 A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

## Join Plot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

## Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex',height=4)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg',height=4)
```

### Pairplot

```python
# When comparing more than two features we use pairplot, it does permutation and combination of all the features
sns.pairplot(df,height=2)
```

```python
sns.pairplot(df,height=2,hue='sex')
```

## DistPlot
It helps us to check the distribution of the column features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Features
 Seaborn also helps us in doing the analysis on Categorical data points. 

 #### -Boxplot
 #### -violinplot
  #### -countplot
  #### -bar plot



```python
# Countplot
sns.countplot(x='sex',data=df)
```

```python
# Countplot
sns.countplot(x='day',data=df,)
```

```python
# Barplot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x="total_bill",y='day',data=df,hue='smoker')
```

## Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernal density estimation and the box plot

```python
sns.violinplot(x="total_bill",y="day",data=df,hue='smoker',palette='rainbow')
```
