### MatplotLib 
#### Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+
#### Some of the major Pros of Matplotlib are:
##### • Generally easy to get started for simple plots
##### • Support for custom labels and texts
##### • Great control of every element in a figure
##### • High-quality output in many formats
##### • Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
#  writing %matplotlib line, so that after every command we need not to write plt.show(), to see the plot
```

```python
import numpy as np
```

```python
#Simple Examples
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
# plotting using matplotlib
# plt scatter
# c stands for color
plt.scatter(x,y,c='b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
# now this figure will be saved in ur file working location
```

```python
y=x*x
```

```python
# plt plot
plt.plot(x,y,marker='o',linestyle='dashed',linewidth=3,markersize=8)
# shift tab for more ingo in brackets above
```

```python
# plt plot
plt.plot(x,y,'ro-')
# shift tab for more ingo in brackets above
```

```python
# plt plot
plt.plot(x,y,'r*',linestyle='dotted',markersize=12)
# shift tab for more ingo in brackets above
# '-' or 'solid': Solid line (default)
# '--' or 'dashed': Dashed line
# '-.' or 'dashdot': Dash-dot line
# ':' or 'dotted': Dotted line
```

```python
# creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'b')

plt.subplot(2,2,2)
plt.plot(x,y,'g')

plt.subplot(2,2,3)
plt.plot(x,y,'r')

plt.subplot(2,2,4)
plt.plot(x,y,'y',linestyle='--')
```

```python
x=np.arange(2,26,2)
y=x*2+4
plt.plot(x,y)
plt.title('Matplotlib Demo Graph')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
```

```python
# SINE WAVE FORM
x=np.arange(0,4*np.pi,0.2)
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.title('Sine Wave Form')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
```

```python
# Subplot
x=np.arange(2,8*np.pi,0.2)
sin_y=np.sin(x)
cos_y=np.cos(x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,sin_y,'--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,cos_y,'g*')
plt.title('Cosine')
```

```python
# Bar Plot
# Comparison
x1=[1,3,6]
y1=[3,7,8]

x2=[12,13,24]
y2=[12,17,30]

plt.bar(x1,y1,color='g')
plt.bar(x2,y2)

```

```python
# Histogram
# Withrespect to numbers what is the density or the count
a=np.array([1,23,45,6,37,23,34,45,23,21,60])
plt.hist(a,bins=12)
plt.title('Histogram')
plt.ylabel('Density or Count')
```

```python
# BOX PLOT USING MATPLOTLIB
# Box plot helps u to find out percentiles, very important in Statistics
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,patch_artist=True,vert=True)
```

### Pie Chart

```python
labels='Python','ML','Data Analytics','Visualisation'
sizes=[30,25,35,10]
explode=(0.2,0,0,0)
colors=['pink','lightblue','green','yellow']
plt.pie(sizes,labels=labels,explode=explode,shadow=True,autopct='%1.1f%%',colors=colors)
```
